```python
level_enco=LabelEncoder()
data["species"]=level_enco.fit_transform(data["species"])
data.head(11)
```

```python
x=data.drop(columns="species")
y=data["species"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
LR=LogisticRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
KNN=KNeighborsClassifier()
```

```python
KNN.fit(x_train,y_train)
```

```python
DT= DecisionTreeClassifier()
```

```python
DT.fit(x_train,y_train)
```

```python
LR_accuracy=LR.score(x_train,y_train)
KNN_accuracy=LR.score(x_train,y_train)
DT_accuracy=LR.score(x_train,y_train)
```

```python
print(f"accuracy by using Logistic Regression: {LR_accuracy}%")
```

```python
print(f"accuracy by using k Nearwst neighbours algorithm: {KNN_accuracy}%")
```
